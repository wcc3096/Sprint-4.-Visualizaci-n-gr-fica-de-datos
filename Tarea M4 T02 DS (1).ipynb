A la hora de visualizar los datos en raw, nos damos cuenta qeu no tienen cabecera, por lo cual crearemos los valores para cada columna

```python
head = ['id', 'movie - year', 'gen']
movies = pd.read_csv('https://raw.githubusercontent.com/BIDDL3/PRE-PROCESSING-DATA/main/movies.dat', 
                     sep='::',header = None,names = head,engine='python', encoding='latin-1')
movies.head()
```

Podemos visualizar que en las 2 variables categ√≥ricas que tenemos , cada fila cuenta con m√°s de un item en c/u de las variables. 

Empezaremos extrayendo los valores de Year y creamos una variable nueva con dicho nombre.

```python
movies['movie'] = movies['movie - year'].apply(lambda x: pd.Series(str(x).rsplit("(", 1)[0].strip()))
movies['year'] = movies['movie - year'].apply(lambda x: pd.Series(str(x).rsplit("(", 1)[1].strip(')')))
movies['year'] = movies['year'].astype(int)
del(movies['movie - year'])
```

```python
movies
```

```python
movies['gen'] = movies.gen.str.split('|')
movies.head(10)
```

```python
movies['gen'] = movies.gen.str[0].str.strip()
movies.head(10)
```
